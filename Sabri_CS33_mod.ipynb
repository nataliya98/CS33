## 10. For each university, find the number of rankings in which they appear (it suffices to appear in one year for each ranking).

```python
rank1 = pd.DataFrame(cwur.groupby('institution', as_index = False).count()[['institution','year']])
rank1['cwur'] = 1
rank1
```

```python
rank2 = pd.DataFrame(shangai.groupby('university_name', as_index = False).count()[['university_name','year']])
rank2['shangai'] = 1
rank2
```

```python
rank3 = pd.DataFrame(times.groupby('university_name', as_index = False).count()[['university_name','year']])
rank3['times'] = 1
rank3
```

```python
rank12 = pd.merge(rank1, rank2, left_on = 'institution', right_on = 'university_name', how = 'outer')
rank123 = pd.merge(rank12, rank3, left_on = 'institution', right_on = 'university_name', how = 'outer')
rank123
```

```python
rank123['cwur'].replace(np.nan, 0, inplace=True)
rank123['shangai'].replace(np.nan, 0, inplace=True)
rank123['times'].replace(np.nan, 0, inplace=True)
rank123
```

```python
rank123['numer_rank'] = rank123['cwur'] + rank123['shangai'] + rank123['times']
rank123
```
