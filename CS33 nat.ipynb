## 5. Consider only the most recent data point of the "times" dataset. Compute the number of male and female students for each country.

```python
times.head()
```

```python
times.head()
```

```python
times['female_%']= times['female_male_ratio'].str.extract('(\d+)').astype(float)
times.head()
```

```python
country = times.iloc[times.groupby('country').idxmax()['year']][['country','year','num_students','female_%']]
country
```

```python
country['female_stud'] = (country['num_students'] *(country['female_%']/100)).round(0)
country
```

```python
country['male_stud'] = country['num_students'] - country['female_stud']
country
```

## 6. Find the universities where the ratio between female and male is below the average ratio (computed over all universities)

```python
times.head()
```

```python
times['ratio'] = (times['female_%']/100)/(1-times['female_%']/100)
times
```

```python
times[times['ratio']== np.inf]
```

```python
times['ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
ratio= times['ratio'].mean()
ratio
```

```python
university_ratio=times[times['ratio']<=ratio]
university_ratio['university_name'].nunique
```

## 7. For each country, compute the fraction of the students in the country that are in one of the universities computed in the previous point (that is, the denominator of the ratio is the total number of students over all universities in the country).


```python
university_ratio
```

```python
times1 = times.groupby('country', as_index=False).sum()[['country','num_students']]
times1
```

```python
df= pd.merge(university_ratio, times1, on = 'country')
df= df.rename(columns={"num_students_y":"tot_students_ country"})
df
```

```python
df['fraction']= df['num_students_x']/df['tot_students_ country']*100
df
```

## un altro metodo

```python
def fraction(stud):
    for n in times1['country']:
        if n == stud['country']:
            return (stud['num_students']/times1['num_students']*100)
```

```python
university_ratio['fraction']= university_ratio.apply(fraction, axis=1)
university_ratio
```
