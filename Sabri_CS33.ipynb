```python
indice = ['1985', '1986', '1987', '1990', '1991',
       '1992', '1993', '1995', '1996', '1997', '1998', '1999', '2000', '2001',
       '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010',
       '2011', '2012', '2013', '2015']
```

```python
a = riga1['country_name']
b = riga1['series_name']
for i in indice:
    if not pd.isna(riga1[i]):
        e = riga1[i]
        dfx = pd.Series([a, b, i, e], index = ['country_name', 'series_name', 'year', 'value'])
        df_lenght = len(my_baby)
        my_baby.loc[df_lenght] = dfx
```

```python
def my_precious(riga):
    a = riga['country_name']
    b = riga['series_name']
    for i in indice:
        if not pd.isna(riga[i]):
            e = riga[i]
            dfx = pd.Series([a, b, i, e], index = ['country_name', 'series_name', 'year', 'value'])
            df_lenght = len(my_baby)
            my_baby.loc[df_lenght] = dfx
    return my_baby
```

10. For each university, find the number of rankings in which they appear (it suffices to appear in one year for each ranking).

```python
rank1 = pd.DataFrame(cwur.groupby('institution', as_index = False).count()[['institution','year']])
rank1['cwur'] = True
rank1
```

```python
rank2 = pd.DataFrame(shangai.groupby('university_name', as_index = False).count()[['university_name','year']])
rank2['shangai'] = True
rank2
```

```python
rank3 = pd.DataFrame(times.groupby('university_name', as_index = False).count()[['university_name','year']])
rank3['times'] = True
rank3
```

```python
rank12 = pd.merge(rank1, rank2, left_on = 'institution', right_on = 'university_name', how = 'outer')
rank123 = pd.merge(rank12, rank3, left_on = 'institution', right_on = 'university_name', how = 'outer')
rank123
```
